```python
from random import shuffle
import pandas as pd
```

```python
def generate_groups(leads: list, members: list):
    shuffle(leads)
    shuffle(members)
    final_teams = {lead: [] for lead in leads}
    while(len(members) > 0):
        shuffle(leads)
        for lead in leads[:len(members)]:
            final_teams[lead].append(members.pop())
    df = pd.DataFrame.from_dict({k: pd.Series(v)
                                 for k, v in final_teams.items()}, orient='index')
    df.reset_index(level=0, inplace=True)
    return df.fillna('')
```
